# House of Pies

In this activity, you will build an order form that displays a list of pies and then prompts users to make a selection. The form will continue to prompt for selections until the user decides to end the process.

### Instructions

* Create an order form that displays a list of pies to the user in the following way:


```
Welcome to the House of Pies! Here are our pies:

---------------------------------------------------------------------
(1) Pecan, (2) Apple Crisp, (3) Bean, (4) Banoffee,  (5) Black Bun, (6) Blueberry, (7) Buko, (8) Burek,  (9) Tamale, (10) Steak
```

* Then, prompt the user to enter the number for the pie they would like to order.

* Immediately follow up their order with `Great! We'll have that <PIE NAME> right out for you`, and then ask if they would like to make another order. If so, repeat the process.

* Once the user is done purchasing pies, print the total number of pies ordered.

````python
def PlacePieOrder(menu):
    
    import collections
    
    # instantiate an empty pie order and keep shopping indicator for the user
    pie_order = []
    still_shopping = 'y'
    
    while still_shopping == 'y':
        new_pie = input("Please enter the menu number for the pie you'd like to order: ")
        
        if not (new_pie.isdigit() and int(new_pie) in range(1, len(menu) + 2)):
            print("Hmm, that's not gonna work. I need an order from the menu, friend.")
            continue
            
        else:
            new_pie = int(new_pie)
            pie_order.append(menu[new_pie])
            
            cart_sum = collections.Counter(pie_order)
            
            print(f"Great! I added a {menu[new_pie]} pie to your order.\n" + 
                 f"Your order looks like this:\n\n")
            for item in cart_sum: print(f"{item} — {cart_sum[item]}")
            
            
            still_shopping = input("Would you like to keep shopping (y/n)?")
            
            while still_shopping not in ('y', 'n'):
                still_shopping = input("Sorry, I don't understand. Would you like to keep shopping (y/n)?")
        
    print("Alright, then, your final order looks like this: \n\n")
    for item in cart_sum: print(f"{item} — {cart_sum[item]}")
    
    return pie_order
````

````python
def StartShopping():
    
    # load the pie menu
    menu = {
        1 : 'Pecan',
        2 : 'Apple Crisp',
        3 : 'Bean',
        4 : 'Banoffee',
        5 : 'Black Bun',
        6 : 'Blueberry',
        7 : 'Buko',
        8 : 'Burek',
        9 : 'Tamale',
        10 : 'Stea'
    }
    
    # display welcome message and menu
    print("Welcome to the House of Pies! Here are our pies:\n\n" + 
         "---------------------------------------------------------------------\n")
    for pie in menu: print(f"({pie}) {menu[pie]}")
    
    # instantiate a still shopping indicator for the user
    strt = input("Would you like to start an order (y/n)?")
        
    while strt not in ('y', 'n'):
        strt = input("Sorry, I don't understand. Would you like to start an order?")
        
    if strt == 'n':
        print("Okay, no worries! Hope to serve you up a slice soon. <3")
        
    else:
        PlacePieOrder(menu)
        
        
    return
````

````python
StartShopping()
````
