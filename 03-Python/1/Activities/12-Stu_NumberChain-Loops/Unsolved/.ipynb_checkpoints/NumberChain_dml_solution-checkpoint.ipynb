# Number Chain

In this activity, you will take user input and print out a string of numbers.

### Instructions

* Using a `while` loop, ask the user "How many numbers?", and then print out a chain of numbers in increasing order, from 0 to the user-input number.

* After the results have been printed, ask the user if they would like to continue.

    * If "y" is entered, keep the chain running by inputting a new number and starting a new count from 0 to the new user-input number.

    * If "n" is entered, exit the application.

```python
# define a function that handles the entire task
def HowMany():
    
    # instantiate a user input variable with an empty str
    user_input = ''
    
    # instantiate a continue variable
    cont = 'y'
    
    while cont == 'y':
        
        # take new user input
        user_input = input("How many numbers?")
        
        # validate user input as int, then complete the print task
        if not user_input.isdigit():
            continue
        
        else:
            for i in range(int(user_input) + 1): print(i)
        
        # ask the user if they'd like to continue
        cont = input("Would you like to continue (y/n)?")
        
        # validate user input for cont variable
        while cont not in ('y', 'n'):
            cont = input ("Sorry, I didn't understand that. Would you like to continue (y/n)?")
    
    return
```

```python
# call our function
HowMany()
```

#### Bonus

Rather than just displaying numbers starting from 0, have the numbers begin at the end of the previous chain.

```python
# define a print function for various cases
def PrintNums(prev, cur):
    
    # case where previous and current inputs match
    if prev == cur:
        print("We're already there!")
    
    # case where current user input is smaller than previous input
    elif prev > cur:
        for i in reversed(range(cur, prev +1)): print(i)
    
    # case where current user input is larger than previous input
    else:
        for i in range(prev, cur +1): print(i)
    
    return
```

```python
def GetTarget():
    
    user_input = input("Where do we want to count to?")
    
    while not user_input.lstrip('-').isdigit():
        user_input = input("Sorry, I need an integer. Where do we want to count to?")
        
    return int(user_input)
```

```python
# define a workhorse function
def ShowMeTheNumbers():
    
    prev_input = 0
    cont = 'y'
    
    while cont == 'y':
        
        cur_input = GetTarget()
        PrintNums(prev_input, cur_input)
            
        cont = input("Would you like to continue (y/n)?")
        
        while cont not in ('y', 'n'):
            cont = input("Sorry, I didn't understand that. Would you like to continue (y/n)?")
            
        if cont == 'n':
            continue
        
        else:
            prev_input = cur_input
```

```python
ShowMeTheNumbers()
```
