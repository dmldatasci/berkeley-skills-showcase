```python
# Dependencies
import requests
from config import api_key
api_key = "fvAG7OyUfSzjkrWifbGgAIG374piUiJN"

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
query = "soccer"


# Search for articles published between a begin and end date
begin_date = "20220101"
end_date = "20220110"


# Build url
# query_url = url + "api-key=" + nyt_api_key + "&q=" + query
query_url = f"{url}api-key={api_key}&q={query}&begin_date={begin_date}&end_date={end_date}&page="
query_url
```

```python
# Retrieve articles
all_articles = []
for i in range(10):
    articles = requests.get(f"{query_url}{i}").json()
    all_articles.append(articles["response"]["docs"])
    
```
