```python
# load dependencies
import pandas as pd
```

```python
# instantiate a data dictionary
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["John", "Kelly", "Sam", "April", "Bobbo"],
    "email": ["jman@gmail", "kelly@aol.com", "sports@school.edu", "April@yahoo.com", "HeyImBobbo@msn.com"]
}

# create a dataframe
customer_df = pd.DataFrame(raw_data_info, columns=["customer_id", "name", "email"])

# display
customer_df
```
