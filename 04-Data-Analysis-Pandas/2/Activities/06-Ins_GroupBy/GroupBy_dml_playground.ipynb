```python
# load dependencies
import pandas as pd
```

```python
# create a reference the CSV file desired
csv_path = "Resources/CT_fires_2015.csv"

# Read the CSV into a Pandas DataFrame
fires_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
fires_df.head()
```

```python
# visualize the list of dataframe columns
list(fires_df.columns)
```

```python
# rename the misspelled column "Propery Loss"
fires_df = fires_df.rename(columns={"Propery Loss" : "Property Loss"})

list(fires_df.columns)
```

```python
# reduce the dataframe to the following columns:
# Fire Department Name, Incident date, Incident Type Code, Incident Type,
# Alarm Date and Time, Arrival Date and Time, Last Unit Cleared Date and Time, 
# Property Loss, Contents Loss, Fire Service Deaths, Fire Service Injuries, 
# Other Fire Deaths, Other Fire Injuries, Incident City, Incident Zip Code
fires_df_reduced = fires_df[["Fire Department Name", "Incident date", "Incident Type Code",
                          "Incident Type", "Alarm Date and Time", "Arrival Date and Time", 
                          "Last Unit Cleared Date and Time", "Property Loss", "Contents Loss", 
                          "Fire Service Deaths", "Fire Service Injuries", "Other Fire Deaths", 
                          "Other Fire Injuries", "Incident City", "Incident Zip Code"]]

fires_df_reduced.head()
```

```python
# check for missing data
fires_df_reduced.count()

# it looks like the missing data is all in numerical columns, 
# and it appears to make sense to use 0s in place of NAs
fires_filled = fires_df_reduced.fillna(0)

# check to ensure missing data was filled
fires_filled.count()
```

```python
# ensure missing data was filled with zeros
fires_filled.head()
```

```python
# ensure that we are working with datetime data in the appropriate columns
# start by making a copy
fires_clean = fires_filled.copy()

# and then convert the appropriate columns to the correct data types
fires_clean = fires_clean.astype({
    "Alarm Date and Time" : "datetime64",
    "Arrival Date and Time" : "datetime64",
    "Last Unit Cleared Date and Time" : "datetime64"
})

fires_clean.head()
```

```python
# use our newly minted datetime objects to calculate response time in seconds
fires_clean["Response Time (s)"] = (fires_clean["Arrival Date and Time"] - fires_clean["Alarm Date and Time"]).dt.total_seconds().astype("int")

# use our newly minted datetime objects to calculate incident duration in seconds
fires_clean["Incident Duration (s)"] = (fires_clean["Last Unit Cleared Date and Time"] - fires_clean["Alarm Date and Time"]).dt.total_seconds().astype("int")

fires_clean.head()
```

```python
# put cities in title case
fires_clean["Incident City"] = fires_clean["Incident City"].str.title()
```

```python
# get a count for how many total incidents have occurred in every city
incident_count_city = fires_clean["Incident City"].value_counts()

incident_count_city
```

```python
# we only want incidents of loss, so we can filter the dataframe
fires_loss = fires_clean.loc[(fires_clean["Property Loss"] > 0) | (fires_clean["Contents Loss"] > 0), :]

fires_loss.head()
```

```python
# get a loss incident count for each city
loss_incidents_city = fires_loss["Incident City"].value_counts()

loss_incidents_city
```

```python
# explore the groupby method to count how many loss incidents occurred in each city
losses_by_city = fires_loss.groupby(["Incident City"])

# notice that we have a `DataFrameGroupBy` object now
print(losses_by_city)
```

```python
# we can use `count()` to see how many loss incidents occurred in each city
losses_by_city.count().head()
```

```python
# we can be pickier and select some numerical columns and use the `sum()` method to get aggregations across fields
losses_by_city[["Property Loss", "Contents Loss"]].sum()
```

```python
# or we can ask for another type of aggregation
losses_by_city[["Property Loss", "Contents Loss"]].mean()
```

```python
# we can also create a new dataframe by saving interesting information as Series and defining a new data structure
tot_property_loss_city = losses_by_city["Property Loss"].sum()
tot_contents_loss_city = losses_by_city["Contents Loss"].sum()

avg_property_loss_city = losses_by_city["Property Loss"].mean()
avg_contents_loss_city = losses_by_city["Contents Loss"].mean()

city_loss_summary = pd.DataFrame({
    "Number of Loss Incidents" : loss_incidents_city,
    "Total Property Loss" : tot_property_loss_city,
    "Average Property Loss" : avg_property_loss_city,
    "Total Contents Loss" : tot_contents_loss_city,
    "Average Contents Loss" : avg_contents_loss_city
})

city_loss_summary
```
