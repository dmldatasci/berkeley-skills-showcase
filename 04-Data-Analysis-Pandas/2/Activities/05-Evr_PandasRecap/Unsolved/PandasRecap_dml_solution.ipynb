```python
# load dependencies
import pandas as pd
```

```python
# create a reference to the `CT_fires_2015.csv` file
filepath = "../Resources/CT_fires_2015.csv"

# read the CSV into a pandas dataframe
ct_fires_raw = pd.read_csv("Resources/CT_fires_2015.csv")

# Print the first five rows of data to the screen
ct_fires_raw.head()
```

```python
# list(df.columns)
list(ct_fires_raw.columns)
```

```python
#df2 = ct_fires_raw.count().to_frame().reset_index()
#df2.columns = ["index", "val"]
#df2

ct_fires = ct_fires_raw.count().to_frame().reset_index()
ct_fires.columns = ["index", "val"]
ct_fires
```

```python
# check the names of all the columns and see if there are any rows with missing data
ct_fires = ct_fires_raw.count().to_frame().reset_index()
ct_fires.columns = ["index", "val"]

ct_fires_count = (len(ct_fires_raw)/100) * 80
display(ct_fires.head())
ct_fires_count
```

```python
for i, row in ct_fires.iterrows():
    if row["val"] < ct_fires_count:
        print(i, row["index"], row["val"])
```

```python
# rename mistyped columns "Aid Given or Received Code " and "Propery Loss"
ct_fires_clean = ct_fires_raw.rename(columns={
    "Aid Given or Received Code " : "Aid Given or Received Code",
    "Propery Loss" : "Property Loss"
})

list(ct_fires_clean.columns)
```

```python
# reduce to columns: 
# Reporting Year, Fire Department Name, Incident date, Incident Type,
# Aid Given or Received Code, Aid Given or Received, Number of Alarms, Alarm Date and Time,
# Arrival Date and Time, Last Unit Cleared Date and Time, Actions Taken 1, Actions Taken 2,
# Actions Taken 3, Property Value, Property Loss, Contents Value, Contents Loss,
# Fire Service Deaths, Fire Service Injuries, Other Fire Deaths, Other Fire Injuries,
# Property Use, Incident Street Address, Incident Apartment Number, Incident City, Incident Zip Code

ct_fires_clean = ct_fires_clean[["Reporting Year", "Fire Department Name", "Incident date", "Incident Type",
                                 "Aid Given or Received Code", "Aid Given or Received", "Number of Alarms", 
                                 "Alarm Date and Time", "Arrival Date and Time", "Last Unit Cleared Date and Time", 
                                 "Actions Taken 1", "Actions Taken 2", "Actions Taken 3", "Property Value", 
                                 "Property Loss", "Contents Value", "Contents Loss", "Fire Service Deaths", 
                                 "Fire Service Injuries", "Other Fire Deaths", "Other Fire Injuries", "Property Use", 
                                 "Incident Street Address", "Incident Apartment Number", "Incident City" , 
                                 "Incident Zip Code"]]

ct_fires_clean.head()
```

```python
# fill NAs for columns "Actions Taken 1", "Actions Taken 2", "Actions Taken 3", 
# and "Incident Apartment Number" with ''
ct_fires_clean[["Actions Taken 1", "Actions Taken 2", "Actions Taken 3", "Incident Apartment Number"]] = ct_fires_clean[["Actions Taken 1", "Actions Taken 2", "Actions Taken 3", "Incident Apartment Number"]].fillna("")


# fill NAs for columns "Other Fire Deaths", "Other Fire Injuries",
# "Property Loss", and "Contents Loss" with 0
ct_fires_clean[["Other Fire Deaths", "Other Fire Injuries", "Property Loss", "Contents Loss"]] = ct_fires_clean[["Other Fire Deaths", "Other Fire Injuries", "Property Loss", "Contents Loss"]].fillna(0)


# better syntax
ct_fires_clean = ct_fires_clean.fillna({"Actions Taken 1" : "", 
                                        "Actions Taken 2" : "", 
                                        "Actions Taken 3" : "",
                                        "Incident Apartment Number" : "",
                                        "Other Fire Deaths" : 0,
                                        "Other Fire Injuries" : 0, 
                                        "Property Loss" : 0,
                                        "Contents Loss" : 0
                                       })

# visualize to ensure it worked
ct_fires_clean[["Actions Taken 1", "Actions Taken 2", "Actions Taken 3", "Incident Apartment Number",
                "Other Fire Deaths", "Other Fire Injuries", "Property Loss", "Contents Loss"]].head()
```

```python
# check counts
ct_fires_clean.count()
```

```python
# remove remaining rows with missing data
ct_fires_clean = ct_fires_clean.dropna(how='any')

ct_fires_clean.count()
```

```python
# incident city needs some cleaning
ct_fires_clean["Incident City"] = ct_fires_clean["Incident City"].str.title()

ct_fires_clean["Incident City"]
```

```python
# filter data to incidents that caused Property or Contents Loss
ct_fires_loss = ct_fires_clean.loc[(ct_fires_clean["Property Loss"] > 0) | (ct_fires_clean["Contents Loss"] > 0), :]

ct_fires_loss.head()
```

```python
# count how many incidents occured in each city
incident_count_city = ct_fires_loss["Incident City"].value_counts()

incident_count_city
```

```python
# convert the city_counts Series into a DataFrame
city_incident_df = pd.DataFrame(incident_count_city)

city_incident_df.head()
```

```python
# convert the column name into "Sum of Loss Incidents"
city_incident_df = city_incident_df.rename(columns={"Incident City" : "Sum of Loss Incidents"})

city_incident_df.head()
```

```python
# calculate the number of deaths from fire incidents where loss occurred
tot_deaths = (ct_fires_loss["Fire Service Deaths"] + ct_fires_loss["Other Fire Deaths"]).sum()
tot_deaths
```

```python
# want to calculate fire department response times, but if we look at the data types fo relevant columns...
# we see that we have an object (which is a string in the case of the dates)
ct_fires_loss.dtypes
```

```python
# convert relevant date columns to datetime
ct_fires_loss = ct_fires_loss.astype({"Alarm Date and Time" : "datetime64",
                                      "Arrival Date and Time" : "datetime64",
                                      "Last Unit Cleared Date and Time" : "datetime64"})

ct_fires_loss.dtypes
```

```python
# now it is possible to find the response time in seconds
# create a new column for "Response Time (seconds)" and use .dt.total_seconds() to calculate the seconds
ct_fires_loss["Response Time (s)"] = (ct_fires_loss["Arrival Date and Time"] - ct_fires_loss["Alarm Date and Time"]).dt.total_seconds().astype("int")

ct_fires_loss.head()
```

#### Hints

* A list of a DataFrame's data types can be checked by accessing its `dtypes` property.

* To change a non-numeric column to a numeric column, use the `df.astype(<datatype>)` method and pass in the desired data type as the parameter.
