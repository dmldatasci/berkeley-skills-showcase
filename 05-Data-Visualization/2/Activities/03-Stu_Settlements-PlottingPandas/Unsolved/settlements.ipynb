```python
# Dependencies
import pandas as pd
```

```python
df = pd.read_csv("Resources/union_settlements_1995.csv")
df.head()
```

```python
# Read CSV
```

```python
# Get total settlements by union
union_data = df['UNION'].value_counts()
union_data
```

```python
union_plot = union_data.plot(kind="bar", facecolor="red"
                , title="Major Settlements"
                , xlabel="Union"
                , ylabel="Settlements"
               )

union_plot.set_xticklabels(union_data.index, rotation=45, rotation_mode="anchor", ha="right", wrap=True)
```
